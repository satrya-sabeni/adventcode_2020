```python
with open('input/input_day9.txt') as file:
    preamble = file.read().split()
```

```python
preamble = [int(x) for x in preamble] 
```

```python
start = 0
current_idx = 25

while current < 1000:
    
    print(f'testing {preamble[current]}...')
    
    area = preamble[start:current_idx]
    found = False
    
    for i in range(len(area)):
        check_ind = area[i]
        to_check = area[0:i] + area[i+1:len(area)]
        for j in to_check:
            if check_ind + j == preamble[current]:
                found = True
        
        if found == True:break
    
    if found == False:
        print(f'{preamble[current_idx]}: is not solvable')
        break 
    
    start += 1
    current += 1
            
```
