```python
with open('input/input_day5.txt') as file:
    seats = file.read().splitlines()
```

```python
seats[:10]
```

```python
seat_ids = []

for seat in seats:
    
    start = 0
    end = 127
    
    for k, rc in enumerate(seat[:7]): #seat[:7]:
        if k is not 6:
            if rc == 'B':
                start = int(start + ((end - start) / 2) + 1)
                end = end
            else:
                start = start
                end = int(start + ((end-start) / 2)) 
        else:
            if rc == 'B':
                row = end
            else:
                row = start
        
        #print(f'seat {rc}:{start} - {end}')
        
    # process column
    start = 0
    end = 7
        
    for cc in seat[7:9]:

        if cc == 'R':
            start = int(start + ((end - start) / 2) + 1)
            end = end
        else:
            start = start
            end = int(start + ((end-start) / 2))
    
    if seat[-1] == 'R':
        col = end
    else:
        col = start

        #print(f'column {cc}:{start} - {end}')
    
    seat_id = row * 8 + col
    
    print(f'seat: row:{row} | column:{col} | id:{seat_id}')
    
    seat_ids.append(seat_id)
            
```
