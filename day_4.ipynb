```python
import re
```

```python
with open('input/input_day4.txt') as file:
    passports = file.read().splitlines()
```

```python
passports
```

```python
passports.append('') # because last got newline didn't for included
```

```python
pass_deviders = [k for k, p in enumerate(passports) if p == '']
```

```python
pass_deviders[-10:]
```

```python
passport_list = []
pass_start = 0
for p in pass_deviders:
    passport = passports[pass_start:p]
    passport_attributes = [a.split(' ') for a in passport]
    passport_processed = sum([attr for attr in passport_attributes], [])
    passport_list.append(passport_processed)
    pass_start = p + 1
```

```python
len(passport_list)
```

```python
passport_dictlist = []
for passport in passport_list:
    passdict = {a.split(':')[0]:a.split(':')[1] for a in passport}
    passport_dictlist.append(passdict)
```

```python
checks = ['ecl', 'pid', 'eyr', 'hcl', 'byr', 'iyr', 'hgt'] # 'cid optional'
```

```python
complete_files = []
incomplete_files = []

complete_count = 0
for passp in passport_dictlist:
    if all([(i in passp) for i in checks]) == True: #(len(passp) >= 7) and ('cid' in passp):
        complete_count += 1
        complete_files.append(passp)
    else:
        incomplete_files.append(passp)
```

```python
# deep passport validator
```

```python
valid_passports = 0
valid_docs = []

for passp in complete_files:
    
    # check birth
    byear = int(passp['byr'])
    if (byear < 1920) or (byear > 2002):
        continue
    
    # check issue year
    iyr = int(passp['iyr'])
    if (iyr < 2010) or (iyr > 2020):
        continue
    
    # check exp date
    eyr = int(passp['eyr'])
    if (eyr < 2020) or (eyr > 2030):
        continue
    
    # check hgt
    if ('cm' in passp['hgt']) or ('in' in passp['hgt']):
        if 'cm' in passp['hgt']:
            hgt = int(passp['hgt'].split('cm')[0])
            if (hgt < 150) or (hgt > 193):
                continue
        else:
            hgt = int(passp['hgt'].split('in')[0])
            if (hgt < 59) or (hgt > 76):
                continue
    else:
        continue
    
    # check haircolor
    if (passp['hcl'][0] != '#') or bool(re.search('[g-z]', passp['hcl'])):
        continue
    
    # check eye color
    if passp['ecl'] not in ['amb','blu','brn','gry','grn','hzl','oth']:
        continue
    
    # check pid 
    if len(passp['pid']) != 9:
        continue
    
    valid_passports += 1
    valid_docs.append(passp)
```
