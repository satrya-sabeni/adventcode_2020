```python
bagbook = {}
bag_numbers = {}

for bag in bagsfile:
    bag_name = bag.split(' contain ')[0]
    contains = bag.split(' contain ')[1][:-1]
    con_bags = contains.split(', ')
    con_quant = [int(bag[0]) for bag in con_bags if bag != 'no other bags']
    con_bags = [bag[2:] for bag in con_bags if bag != 'no other bags']
    con_bags = list(map(lambda x: x+'s' if x[-1] != 's' else x, con_bags))
    
    
    bagbook[bag_name] = con_bags
    bag_numbers[bag_name] = con_quant
```
